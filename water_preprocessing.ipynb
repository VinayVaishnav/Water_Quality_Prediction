```python
sourceslist = list(np.unique(dfcopy.Source))
sourceslist.sort()
```

```python
sourceslist
```

```python
replacesourceslist = [each for each in range(len(sourceslist))]
replacesourceslist
```

```python
colors = {
    'Colorless': 0,
    'Near Colorless': 1,
    'Faint Yellow': 2,
    'Light Yellow': 3,
    'Yellow': 4,
}
```

```python
months = {
    'January': 1,
    'February': 2,
    'March': 3,
    'April': 4,
    'May': 5,
    'June': 6,
    'July': 7,
    'August': 8,
    'September': 9,
    'October': 10,
    'November': 11,
    'December': 12,    
}
```

```python
colorslist = []
replacecolorslist = []

for each in colors:
    colorslist.append(each)
    replacecolorslist.append(colors[each])
    
print(colorslist)
print(replacecolorslist)
```

```python
monthslist = []
replacemonthslist = []

for each in months:
    monthslist.append(each)
    replacemonthslist.append(months[each])
    
print(monthslist)
print(replacemonthslist)
```

```python
dfcopy['Color'].replace(colorslist, replacecolorslist, inplace=True)
```

```python
dfcopy['Month'].replace(monthslist, replacemonthslist, inplace=True)
```

```python
dfcopy['Source'].replace(sourceslist, replacesourceslist, inplace=True)
```

```python
dfcopy['Day'] = dfcopy['Day'].astype(int)
```

```python
dfcopy['Time of Day'] = dfcopy['Time of Day'].astype(int)
```
